```python
class Document():
    def __init__(self, title, author, context):
        print('init function called')
        self.title = title
        self.author = author
        self.__context = context # __开头的属性是私有属性

    def get_context_length(self):
        return len(self.__context)

    def intercept_context(self, length):
        self.__context = self.__context[:length]

harry_potter_book = Document('Harry Potter', 'J. K. Rowling', '... Forever Do not believe any thing is capable of thinking independently ...')

print(harry_potter_book.title)
print(harry_potter_book.author)
print(harry_potter_book.get_context_length())

harry_potter_book.intercept_context(10)

print(harry_potter_book.get_context_length())

print(harry_potter_book.__context)
```

```python
class YAMLObjectMetaclass(type):
  def __init__(cls, name, bases, kwds):
    super(YAMLObjectMetaclass, cls).__init__(name, bases, kwds)
    print('aha!{}'.format(cls))

class YAMLObject(metaclass=YAMLObjectMetaclass):
  yaml_loader = 'a'

class YAMLObject1(metaclass=YAMLObjectMetaclass):
  yaml_loader = 'a'

YAMLObject()
YAMLObject1()
```

```python
import asyncio

async def main():
    print(f"1. 当前所有 tasks: {len(asyncio.all_tasks())}")
    
    # 直接 await sleep（这是 coroutine，不是 task）
    await asyncio.sleep(0.1)
    print(f"2. await sleep 后 tasks: {len(asyncio.all_tasks())}")
    
    # 先 create_task 再 await
    task = asyncio.create_task(asyncio.sleep(0.1))
    print(f"3. create_task 后 tasks: {len(asyncio.all_tasks())}")
    await print('test')
    
    await task
    print(f"4. await task 后 tasks: {len(asyncio.all_tasks())}")

await main()
```

```python
def CountDown(n):
    while n > 0:
        n -= 1

start = time.perf_counter()
results_single = CountDown(100000000)
single_duration = time.perf_counter() - start
print(f"⏱️  单进程耗时: {single_duration:.2f} 秒\n")
```

```python
from threading import Thread

n = 100000000

t1 = Thread(target=CountDown, args=[n // 2])
t2 = Thread(target=CountDown, args=[n // 2])
t1.start()
t2.start()

start = time.perf_counter()
t1.join()
t2.join()
single_duration = time.perf_counter() - start
print(f"⏱️  单进程耗时: {single_duration:.2f} 秒\n")
```

```python
import threading

n = 0

def foo():
    global n
    n += 1

threads = []
for i in range(100):
    t = threading.Thread(target=foo)
    threads.append(t)

for t in threads:
    t.start()

for t in threads:
    t.join()

print(n)
```

```python
import unittest

# 将要被测试的排序函数
def sort(arr):
    l = len(arr)
    for i in range(0, l):
        for j in range(i + 1, l):
            if arr[i] >= arr[j]:
                tmp = arr[i]
                arr[i] = arr[j]
                arr[j] = tmp


# 编写子类继承unittest.TestCase
class TestSort(unittest.TestCase):

   # 以test开头的函数将会被测试
   def test_sort(self):
        arr = [3, 4, 1, 5, 6]
        sort(arr)
        # assert 结果跟我们期待的一样
        self.assertEqual(arr, [1, 3, 4, 5, 6])

if __name__ == '__main__':
    ## 如果在Jupyter下，请用如下方式运行单元测试
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
    
    ## 如果是命令行下运行，则：
    ## unittest.main()
```

```python
import websocket
import _thread  # Python 3 中 thread 改名为 _thread
import time     # 缺少的导入

def on_message(ws, message):
    print('Received: ' + message)

def on_open(ws):
    def gao():
        for i in range(5):
            time.sleep(0.01)
            msg = "{0}".format(i)
            ws.send(msg)
            print('Sent: ' + msg)
        time.sleep(1)
        ws.close()
        print("Websocket closed")
    
    # 使用 _thread.start_new_thread
    _thread.start_new_thread(gao, ())

if __name__ == "__main__":
    ws = websocket.WebSocketApp("ws://echo.websocket.org/",
                                on_message=on_message,
                                on_open=on_open)
    ws.run_forever()
```
